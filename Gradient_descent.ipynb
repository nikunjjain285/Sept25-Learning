```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2])
y=np.array([300,500])
```

```python
def compute_cost(x,y,w,b):
    m=x.shape[0]
    cost=0
    for i in range(m):
        fx_w_b=w*x[i]+b
        cost=cost+(fx_w_b-y[i])**2
        cost=(cost/2)*m
    return cost
```

```python
def compute_gradient(x,y,w,b):
    df_dw=0
    df_db=0
    m=x.shape[0]
    for i in range(m):
        fx_w_b=w*x[i]+b
        df_dw_i=(fx_w_b-y[i])*x[i]
        df_dw+=df_dw_i
        df_db_i=(fx_w_b-y[i])
        df_db+=df_db_i
    df_dw=df_dw/m
    df_db=df_db/m
    return df_dw,df_db
    
    
```

```python
def gradient_descent(x_train,y_train,w_in,b_in,alpha,num_iter,cost_function,gradient_function):
    J_hist=[]
    p_hist=[]
    w=w_in
    b=b_in
    for i in range(num_iter):
        df_dw,df_db=gradient_function(x_train,y_train,w,b)
        w=w-alpha*df_dw
        b=b-alpha*df_db
        J_hist.append(cost_function(x_train,y_train,w,b))
        p_hist.append([w,b])
    return w,b,J_hist,p_hist
        
```

```python
w_final,b_final,J_hist,p_hist=gradient_descent(x,y,0,0,0.02,10000,compute_cost,compute_gradient)
print(f"(w,b) found by gradient descent : ({w_final:.4f},{b_final:.4f})")
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
ax1.plot(J_hist[:100])
ax2.plot(J_hist[1000:])
ax1.set_title("Cost v/s Iterations [Start]")
ax2.set_title("Cost v/s Iterations [End]")
ax1.set_ylabel("Cost"); ax2.set_ylabel("Cost")
ax1.set_xlabel("Iterations");ax2.set_xlabel("Iterations")
```
